## 2. Data Cleaning

Function to clean Q7 "hat programming languages do you use on a regular basis?", as I would like to have them counted in one single row

```python
import pandas as pd

def finalCount(row):
    count = 0
    if row['Q7_Part_1'] == 'Python':
        count = count + 1 
    if row['Q7_Part_2'] == 'R':
        count = count + 1 
    if row['Q7_Part_3'] == 'SQL':
        count = count + 1
    if row['Q7_Part_4'] == 'C':
        count = count + 1
    if row['Q7_Part_5'] == 'C++':
        count = count + 1 
    if row['Q7_Part_6'] == 'Java':
        count = count + 1
    if row['Q7_Part_7'] == 'Javascript':
        count = count + 1
    if row['Q7_Part_8'] == 'Julia':
        count = count + 1
    if row['Q7_Part_9'] == 'Swift':
        count = count + 1
    if row['Q7_Part_10'] == 'Bash':
        count = count + 1
    if row['Q7_Part_11'] == 'MATLAB':
        count = count + 1
    if row['Q7_Part_12'] == 'None':
        count = count + 1
    if row['Q7_OTHER'] == 'Other':
        count = count + 1
    return count
```

```python
df['programming_lang_c'] = df.apply(finalCount, axis =1)
```

## Data Cleaning

```python
def cleaner(x):
    
    check = ['+','<','>','$',',', ' years', ' employees', ' or more employees']
    
    for i in check:
        if i in str(x):
            x = str(x).replace(i, '')
    return str(x)
```

#### Age column
As the age column contains range of ages, I will use the Ordinal Encoder to encode them in buckets

```python
data['age'] = data['age'].apply(cleaner)
```

As gender columns contains many responses, I will clean it to male, female and others

#### Country column

I will clean long names of countries and replace with the short version

```python
data['country'] = data['country'].replace(['United Kingdom of Great Britain and Northern Ireland', 'United States of America', 'Iran, Islamic Republic of...', 'United Arab Emirates', 'Hong Kong (S.A.R.)','I do not wish to disclose my location'], ['UK', 'USA', 'Iran', 'UAE','Hong Kong','Other'])
```

```python
data['country'].value_counts()[0:60]
```

As my model will focus on the gender pay gap and I have countries from all over the world with very different levels of salaries, I will only consider Europe, USA and India.

```python
def country_group(row):
 
    list_1 = ['UK', 'Russia', 'Germany', 'Spain', 'France', 'Italy', 'Poland', 'Ukraine', 'Netherlands', 'Portugal', 'Greece', 'Ireland', 'Sweden', 'Switzerland', 'Belgium', 'Czech Republic', 'Romania', 'Austria', 'Belarus', 'Denmark'] 
    list_2 = ['USA']
    list_3 = ['India']
    
    if row in list_1: 
        return "Europe"
    elif row in list_2:
        return "USA"
    elif row in list_3:
        return "India"
    else:
        return "Other"
```

```python
data['country_group'] = data['country'].apply(country_group)
```

```python
data['country_group'].value_counts()[0:60]
```

#### Size of company

To reduce my categories, I will group the size of th company in small, medium, large

```python
data['company_size'].value_counts()
```

```python
def size_group(row):
 
    list_1 = ['0-49 employees'] 
    list_2 = ['50-249 employees']
    #list_3 = ['250-999 employees','1000-9,999 employees','10,000 or more employees']
    
    if row in list_1: 
        return "Small"
    elif row in list_2:
        return "Medium"
    else:
        return "Large"
```

#### Role column

As we have too many roles in the survey, I will group them within the three typical jobs categories within the data analytcis area and others:

Dropping roles student and currently not employed as my project focused on gender pay gap and these two roles do not have a salary.

As we have too many roles in the survey, I will group them within the three typical jobs categories within the data analytcis area and others:

| Group          	| Roles grouped                                                                                                    	|
|----------------	|------------------------------------------------------------------------------------------------------------------	|
| Data Analyst   	| Data Analyst, Business Analyst, Program/Project Manager, Product Manager                                         	|
| Data Scientist 	| Data Scientist, Research Scientist, Statistician                                                                 	|
| Data Engineer  	| Data Engineer, DBA/Database Engineer, Software Engineer, Developer Relations/Advocacy, Machine Learning Engineer 	|
| Other          	| Other                                                                                                            	|

```python
def role_group(row):

    
    list_1 = ['Data Analyst', 'Product Manager' ,'Program/Project Manager', 'Business Analyst' ]
    list_2 = ['Data Engineer', 'DBA/Database Engineer','Machine Learning Engineer' ,'Software Engineer', 'Developer Relations/Advocacy']
    list_3 = ['Data Scientist', 'Research Scientist', 'Statistician']
    
    if row in list_1: 
        return "Data Analyst"
    elif row in list_2:
        return "Data Engineer"
    elif row in list_3:
        return "Data Scientist"
    else:
        return "Other"

    
```

```python
data['role_group'] = data['role'].apply(role_group)
```

## Dealing with null values

```python
def check_null(df):
    nulls = pd.DataFrame(df.isna().sum()*100/len(df), columns=['percentage'])
    nulls.sort_values('percentage', ascending = False)
    return nulls
```

```python
check_null(data)
```

As my dataset has a considerable size, I will drop all the null values from columns. For the salary_avg I have decided not to go with the mean of the blanks as I could input wrong information into the data. 

```python
data = data.dropna()
```

```python
check_null(data)
```
